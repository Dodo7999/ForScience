```python
def print_epoch_info(epoch_idx, num_epochs, avg_error, penalty, collate_prints=True):
    with np.printoptions(precision=3):
        # Preserve the first iteration's print, otherwise overwrite.
        if epoch_idx == 0:
            prefix = "Initial performance:\n"
            suffix = "\n"
        elif collate_prints:
            prefix = ""
            suffix = "\n"
        else:
            prefix = ""
            suffix = "\n"
        ndigits = int(np.log10(num_epochs)) + 1
        print(prefix + f"Epoch {epoch_idx+1:{ndigits}d}/{num_epochs} ({epoch_idx/num_epochs:3.0%});"
              f" MSE = {avg_error:.4f};"
              f" penalty = {penalty:.9f}", end=suffix)


MSE_LOSS = torch.nn.MSELoss(reduction="none")

def avg_loss_per_env(model, environments, loss=MSE_LOSS):
    return np.array([float(loss(model(x_e), y_e).mean()) for x_e, y_e in environments])

def eval_dataset(model, dataset_name, dataset_envs, loss=MSE_LOSS):
    env_errors = avg_loss_per_env(model, dataset_envs, loss)
    print(f"{dataset_name} errors:")
    for i, err in enumerate(env_errors):
        print(f"Environment {i+1} Error = {err}")
    print(f"Overall Average = {env_errors.mean():.4f}")

def eval_all(model, loss=MSE_LOSS):
    for name, envs in (("Train", train_environments), ("Test", test_environments)):
        eval_dataset(model, name, envs, loss)


def train(use_IRM=True, num_epochs=1000, epochs_per_eval=1000, collate_prints=True):

    model = nn.Sequential(
            nn.Linear(in_features=125, out_features=64),
            nn.ReLU(),
            nn.Linear(in_features=64, out_features=32),
            nn.ReLU(),
            nn.Linear(in_features=32, out_features=16),
            nn.ReLU(),
            nn.Linear(in_features=16, out_features=1)
    )

    # We will only learn phi.
    opt = torch.optim.Adam(model.parameters(), lr=0.001)
    mse = torch.nn.MSELoss(reduction="none")

    for i in range(num_epochs):
        # Sum of average error in each environment.
        error = 0
        # Total IRM penalty across all environments.
        penalty = 0

        error_e = []
        # Forward pass and loss computation.
        for x_e, y_e in train_environments:
            p = torch.randperm(len(x_e))
            error_e = mse(model(x_e[p]), y_e[p])
            penalty += torch.sum(error_e**2) 
            error += error_e.mean()
        
        error = error / len(train_environments)
        penalty = penalty / len(train_environments)
        
        weight_norm = torch.tensor(0.)
        for w in model.parameters():
            weight_norm += w.norm().pow(2)
        
        # Print losses.
        print_epoch_info(i,
                             num_epochs,
                             error,
                             penalty,
                             collate_prints)
        
        # Backward pass.
        opt.zero_grad()
        if use_IRM:
            # NOTE: This scaling seems arbitrary and I'm not sure if the paper provides insight as
            # to how to choose this hyperparameter.
            (error + 1e-3 * weight_norm + penalty).backward()
        else:
            error.backward()
        opt.step()

        # Do a sanity check.
        if model[0].weight.isnan().any():
            print_epoch_info(i,
                             num_epochs,
                             error / len(train_environments),
                             penalty,
                             collate_prints)
            print("ERROR: Optimization diverged and became NaN. Halting training.")
            return model, model[0].weight
    
    params = model[0].weight
    print("\n\nFinal model:")
    print(f"params = {params.transpose()[0]}")
    eval_all(model, mse)
    return model, model[0].weight
```

```python
class Autoencoder(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(126, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
        )
        self.decoder = nn.Sequential(
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 126),
            nn.ReLU(),
        )

    def forward(self, sample):
        latent = self.encoder(sample)
        reconstructed = self.decoder(latent)
        return reconstructed
```

```python
n_epochs = 20
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print(device)
model1 = Autoencoder().to(device)
optimizer1 = torch.optim.Adam(model1.parameters(), lr=0.001)
loss_fn1 = F.mse_loss
train_losses1 = []
val_losses1 = []

for epoch in tqdm_notebook(range(n_epochs)):
    model1.train()
    train_losses_per_epoch = []
    for i, X_batch in enumerate(train_loader):
        optimizer1.zero_grad()
        #print(X_batch.to(device).float())
        reconstructed = model1(X_batch.to(device).float())
        #print(reconstructed)
        loss = loss_fn1(reconstructed, X_batch.to(device).float())
        loss.backward()
        optimizer1.step()
        train_losses_per_epoch.append(loss.item())

    train_losses1.append(np.mean(train_losses_per_epoch))

    model1.eval()
    val_losses_per_epoch = []
    with torch.no_grad():
        for X_batch in val_loader:
            reconstructed = model1(X_batch.to(device).float())
            loss = loss_fn1(reconstructed, X_batch.to(device).float())
            val_losses_per_epoch.append(loss.item())

    val_losses1.append(np.mean(val_losses_per_epoch))
```

```python
class LinearVAE(nn.Module):
    def __init__(self):
        super(LinearVAE, self).__init__()
 
        # encoder
        self.encoder = nn.Sequential(
            nn.Linear(in_features=126, out_features=64),
            nn.ReLU(),
            nn.Linear(in_features=64, out_features=32)
        )
 
        # decoder 
        self.decoder = nn.Sequential(
            nn.Linear(in_features=16, out_features=32),
            nn.ReLU(),
            nn.Linear(in_features=32, out_features=64),
            nn.ReLU(),
            nn.Linear(in_features=64, out_features=126)
        )

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        sample = torch.normal(mu,std)
        return sample
 
    def forward(self, x):
        x = self.encoder(x).view(-1, 2, 16)
        mu = x[:, 0, :] 
        log_var = x[:, 1, :] 
        z = self.reparameterize(mu, log_var)
 
        x = self.decoder(z)
        reconstruction = torch.arctan(x)
        return reconstruction, mu, log_var

    def sample(self, z):
        generated = self.decoder(z)
        generated = torch.arctan(generated)
        generated = generated.view(-1, 64, 64, 3)
        return generated

    def get_latent_vector(self, x):
        x = self.flatten(x).float()
        x = self.encoder(x).view(-1, 2, 16)
        mu = x[:, 0, :] 
        log_var = x[:, 1, :] 
        z = self.reparameterize(mu, log_var)
        return z
```

```python
def KL_divergence(mu, logsigma):
    loss = -0.5 * torch.sum(1 + logsigma - mu.pow(2) - logsigma.exp())
    return loss

def log_likelihood(x, reconstruction):
    loss = F.mse_loss
    return loss(reconstruction, x)

def loss_vae(x, mu, logsigma, reconstruction):
    return KL_divergence(mu, logsigma) + log_likelihood(x, reconstruction)
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

criterion = loss_vae

autoencoder = LinearVAE().to(device)

optimizer = torch.optim.Adam(autoencoder.parameters(), lr=0.0003)
```

```python
x_train2, x_val2 = train_test_split(X2.astype(float), test_size=0.1, shuffle=False)
train_loader = torch.utils.data.DataLoader(x_train2, batch_size=40)
val_loader = torch.utils.data.DataLoader(x_val2, batch_size=40)
```

```python
n_epochs = 20
train_losses = []
val_losses = []

for epoch in tqdm_notebook(range(n_epochs)):
    autoencoder.train()
    train_losses_per_epoch = []
    for batch in train_loader:
        optimizer.zero_grad()
        reconstruction, mu, logsigma = autoencoder(batch.float().to(device))
        loss = criterion(batch.to(device).float(), mu, logsigma, reconstruction)
        loss.backward()
        optimizer.step()
        train_losses_per_epoch.append(loss.item()/ len(train_loader))

    train_losses.append(np.mean(train_losses_per_epoch))

    autoencoder.eval()
    val_losses_per_epoch = []
    with torch.no_grad():
        for batch in val_loader:
            reconstruction, mu, logsigma = autoencoder(batch.float().to(device))
            loss = criterion(batch.to(device).float(), mu, logsigma, reconstruction)
            val_losses_per_epoch.append(loss.item()/ len(val_loader))

    val_losses.append(np.mean(val_losses_per_epoch))
```
